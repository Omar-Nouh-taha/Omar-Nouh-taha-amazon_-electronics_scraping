```python
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re
import pandas as pd
import time

# Initialize lists
phones = []
prices_phone = []
before_phone = []
num_rate_phone = []
mainphoto_phone = []
about_phone = []
all_reviews_phone = []
ratings_phone = []
p_links_phone = []
descriptions_phone = []

try:
    # Extract phone links from all pages
    for page in page_links_phone:
        driver.get(page)
        
        all_items = driver.find_elements(By.XPATH, '/html/body/div[1]/div[1]/div[1]/div[1]/div/span[1]/div[1]/div[*]/div/div/span/div/div/div[*]/span/a')
        for item in all_items:
            phone_link = item.get_attribute('href')
            p_links_phone.append(phone_link)
    
    # Extract details from each phone link
    for link in p_links_phone:
        driver.get(link)
        
        # Extract phone names
        try:
            phone_name = driver.find_element(By.ID, "productTitle").text
            phones.append(phone_name)
        except NoSuchElementException:
            phones.append("Name not available")

        # Extract phone details (about section)
        try:
            about_phone_elements = driver.find_elements(By.CSS_SELECTOR, ".a-unordered-list.a-vertical.a-spacing-mini .a-spacing-mini")
            phone_details = [item.text for item in about_phone_elements]
            phone_details_str = ', '.join(phone_details)
            about_phone.append(phone_details_str)
        except NoSuchElementException:
            print("About section not found for link:", link)
            about_phone.append(None)
        
        # Extract reviews
        try:
            review_elements = driver.find_elements(By.CLASS_NAME, "a-row.a-spacing-small.review-data")
            reviews = []
            for review_element in review_elements:
                review_text = review_element.find_element(By.TAG_NAME, "span").text
                review_text = review_text.replace('\n', ' ')  
                review_text = review_text.replace('-', '')  
                reviews.append(review_text)
        
            all_reviews_phone.append(reviews)
        except NoSuchElementException:
            print("Review section not found for link:", link)
            all_reviews_phone.append([])

        # Extract description using the specified XPath
        try:
            description_elements = driver.find_elements(By.XPATH, "/html/body/div[*]/div/div[*]/div[*]/div[*]/div[33]/div/div[*]/div/table")
            description_text = ' '.join([elem.text for elem in description_elements])
            descriptions_phone.append(description_text)
        except NoSuchElementException:
            print("Description not found for link:", link)
            descriptions_phone.append("No description")
            
        # Extract price
        try:
            # Wait until the price element is present on the page
            price_after_promo = WebDriverWait(driver, 1).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, 'span.a-price.aok-align-center.reinventPricePriceToPayMargin.priceToPay'))
            )
            
            if price_after_promo:
                # Extract the price text
                price_after_promo_text = price_after_promo.find_element(By.CSS_SELECTOR, 'span[aria-hidden="true"]')
                price_phone = price_after_promo_text.text.strip()

        except (NoSuchElementException, TimeoutException) as e:
            try:
                # Wait for the "see all buying choices" button to be clickable
                see_price_bottom = WebDriverWait(driver, 5).until(
                    EC.element_to_be_clickable((By.ID, "buybox-see-all-buying-choices"))
                )
                see_price_button = see_price_bottom.find_element(By.CLASS_NAME, "a-button-inner")
                see_price_button.click()  # Click the button to view additional prices
            
                # Wait for the new price element to be visible
                WebDriverWait(driver, 5).until(
                    EC.visibility_of_element_located((By.ID, "aod-price-1"))
                )
                
                # Find and extract the price
                p = driver.find_element(By.ID, "aod-price-1").find_element(By.CLASS_NAME, "a-price-whole")
                price_phone = p.text.strip()
                
            except (NoSuchElementException, TimeoutException) as inner_e:
                price_phone = "0.00"  # Default value in case of an error
                
        prices_phone.append(price_phone.replace("EGP", "").replace(",", "").strip())
  
        # Extract price before promo
        try:
            price_before_promo = driver.find_element(By.CSS_SELECTOR, ".a-price.a-text-price")
            price_before_promo_text = price_before_promo.find_element(By.CSS_SELECTOR, "[aria-hidden='true']").text.strip()
            before_phone.append(price_before_promo_text.replace("EGP", "").replace(",", "").strip())
        except NoSuchElementException:
            before_phone.append("No Sale")
        
        # Extract main photo URL
        try:
            main_photo = driver.find_element(By.ID, "imgTagWrapperId").find_element(By.TAG_NAME, "img").get_attribute("src")
            mainphoto_phone.append(main_photo)
        except NoSuchElementException:
            mainphoto_phone.append("Image not found")
        
        # Extract number of ratings
        try:
            number_of_ratings = driver.find_element(By.ID, "acrCustomerReviewText")
            rates_phone = number_of_ratings.text.strip()
            num_rate_phone.append(rates_phone)
        except NoSuchElementException:
            num_rate_phone.append("No Ratings")

        # Extract ratings
        try:
            # Find the element with the ID "acrPopover"
            popover_element = driver.find_element(By.ID, "acrPopover")
            ratings_element = popover_element.find_element(By.CLASS_NAME, "a-size-base.a-color-base")
            ratings_text = ratings_element.text
            ratings_phone.append(ratings_text + " out of 5 stars")
        except NoSuchElementException:
            ratings_phone.append("No Rating")

except Exception as e:
    print(f"An error occurred: {str(e)}")



```

```python
# Create DataFrame
df = pd.DataFrame({
    "Names": phones,
    "Price": prices_phone,
    "Before Promo": before_phone,
    "Main Photo": mainphoto_phone,
    "About Phone": about_phone,
    "All Reviews": all_reviews_phone,
    "Ratings": ratings_phone,
    "Number of Ratings": num_rate_phone,
    "Descriptions": descriptions_phone
})
```

```python
# Extract brands (assuming brand is the first word in the name)
brands = df["Names"].apply(lambda x: x.split(" ")[0])

df['SIM']= df["Names"].apply(lambda x : re.findall("Dual SIM|Dual Sim|Dual|DUAL SIM",x))
sim =df['SIM'].apply(lambda x: ' '.join(x))
df['Ram']=df["Names"].apply(lambda x: re.findall(r'\b(0?[0-9]|1[0-9])GB\b', x))
Ram =df['Ram'].apply(lambda x: ' '.join(x))
df['Storage']=df["Names"].apply(lambda x: re.findall(r'\b(6[1-9]|[7-9][0-9]|[1-9][0-9]{2,})\s*(?:GB|GGb|Storage)\b', x, re.IGNORECASE))
Storage =df['Storage'].apply(lambda x: ' '.join(x))
```

```python
def format_reviews(entry):
    if isinstance(entry, list):
        entry = ','.join(entry)
    parts = entry.split(',')
    formatted_text = '   '.join(f'({part})' for part in parts)
    return formatted_text
df['All Reviews'] = df['All Reviews'].apply(format_reviews)
```

```python
df['Price'] = df['Price'].str.replace(r'\n00', '', regex=True) 
df['Price'] = df['Price'].str.replace(r'\n', '.', regex=True) 
df = df[df['Price'] != "0.00"]
df.reset_index(drop=True, inplace=True)
```

```python
# Function to parse and format the mobile data
def format_description(entry):
    # Check if entry is empty
    if not entry.strip():
        return ''
    
    # Split the entry into key-value pairs
    data_dict = dict(line.split(' ', 1) for line in entry.split('\n') if line)
    
    # Format the dictionary into the desired string format
    formatted_text = '   '.join([f'({key} : {value})' for key, value in data_dict.items()])
    return formatted_text

# Apply the formatting function to the "description" column
df['Descriptions'] = df['Descriptions'].apply(format_description)
```

```python
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re
import pandas as pd
import time
import time
# Initialize lists
screen_protector = []
prices_screen_protector = []
before_screen_protector = []
num_rate_screen_protector = []
mainphoto_screen_protector = []
about_phone_screen_protector = []
all_reviews_screen_protector = []
ratings_screen_protector = []
p_links_screen_protector = []
descriptions_protector = []

try:
    # Extract screen protector links from all pages
    for page in page_links_screen_protector:
        driver.get(page)
        
        # Wait 1 second to allow the page to load
        time.sleep(1)
        
        all_items = driver.find_elements(By.XPATH, '/html/body/div[1]/div[1]/div[1]/div[1]/div/span[1]/div[1]/div[*]/div/div/span/div/div/div[*]/span/a')
        for item in all_items:
            screen_protector_link = item.get_attribute('href')
            p_links_screen_protector.append(screen_protector_link)
    
    # Extract details from each screen protector link
    for link in p_links_screen_protector:
        driver.get(link)
        
        # Extract screen protector names
        try:
            screen_protector_name = driver.find_element(By.ID, "productTitle").text
            screen_protector.append(screen_protector_name)
        except NoSuchElementException:
            screen_protector.append("Name not available")

        # Extract phone details (about section)
        try:
            abouttt_phone = driver.find_elements(By.CSS_SELECTOR, ".a-unordered-list.a-vertical.a-spacing-mini .a-spacing-mini")
            mobile_details = [item.text for item in abouttt_phone]
            mobile_details_str = ', '.join(mobile_details)
            about_phone_screen_protector.append(mobile_details_str)
        except NoSuchElementException:
            print("About section not found for link:", link)
            about_phone_screen_protector.append(None)
        
         # Extract reviews
        try:
            review_elements = driver.find_elements(By.CLASS_NAME, "a-row.a-spacing-small.review-data")
        
            reviews = []
            for review_element in review_elements:
                review_text = review_element.find_element(By.TAG_NAME, "span").text
                review_text = review_text.replace('\n', ' ')  
                review_text = review_text.replace('-', '')  
                reviews.append(review_text)
        
            all_reviews_screen_protector.append(reviews)
   
        except NoSuchElementException:
            print("Review section not found for link:", link)
            all_reviews_screen_protector.append([])
            
       # Extract description using the specified XPath
        try:
            description_elements = driver.find_elements(By.XPATH, "/html/body/div[*]/div/div[*]/div[*]/div[*]/div[33]/div/div[*]/div/table")
            description_text = ' '.join([elem.text for elem in description_elements])
            descriptions_protector.append(description_text)
            
        except NoSuchElementException:
            print("Description not found for link:", link)
            descriptions_protector.append("No description")
            
        
        # Extract price
        try:
            # Wait until the price element is present on the page
            price_after_promo = WebDriverWait(driver, 1).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, 'span.a-price.aok-align-center.reinventPricePriceToPayMargin.priceToPay'))
            )
            
            if price_after_promo:
                # Extract the price text
                price_after_promo_text = price_after_promo.find_element(By.CSS_SELECTOR, 'span[aria-hidden="true"]')
                price_screen_protector = price_after_promo_text.text.strip()

        except (NoSuchElementException, TimeoutException) as e:
            print(f"Primary price extraction failed: {e}")
            try:
                # Click to see additional buying choices if needed
                see_price_bottom = WebDriverWait(driver, 5).until(
                    EC.element_to_be_clickable((By.ID, "buybox-see-all-buying-choices"))
                )
                see_price_button = see_price_bottom.find_element(By.CLASS_NAME, "a-button-inner")
                see_price_button.click()  # Click the button to view additional prices
                
                # Wait for the new price element to be visible
                WebDriverWait(driver, 5).until(
                    EC.visibility_of_element_located((By.ID, "aod-price-1"))
                )
                p = driver.find_element(By.ID, "aod-price-1").find_element(By.CLASS_NAME, "a-price-whole")
                price_screen_protector = p.text.strip()
            except (NoSuchElementException, TimeoutException) as inner_e:
                price_screen_protector = "0.00"  # Default value in case of an error
        
        prices_screen_protector.append(price_screen_protector.replace("EGP", "").replace(",", "").strip())

        # Extract price before promo
        try:
            price_before_promo = driver.find_element(By.CSS_SELECTOR, ".a-price.a-text-price")
            price_before_promo_text = price_before_promo.find_element(By.CSS_SELECTOR, "[aria-hidden='true']").text.strip()
            before_screen_protector.append(price_before_promo_text.replace("EGP", "").replace(",", "").strip())
        except NoSuchElementException:
            before_screen_protector.append("No Sale")
        
        # Extract main photo URL
        try:
            main_photo = driver.find_element(By.ID, "imgTagWrapperId").find_element(By.TAG_NAME, "img").get_attribute("src")
            mainphoto_screen_protector.append(main_photo)
        except NoSuchElementException:
            mainphoto_screen_protector.append("Image not found")
        
        
        # Extract number of ratings
        try:
            number_of_ratings = driver.find_element(By.ID, "acrCustomerReviewText")
            rates_screen_protector = number_of_ratings.text.strip()
            num_rate_screen_protector.append(rates_screen_protector)
        except NoSuchElementException:
            num_rate_screen_protector.append("No Ratings")

        # Extract ratings
        try:
            # Find the element with the ID "acrPopover"
            popover_element = driver.find_element(By.ID, "acrPopover")
            ratings_element = popover_element.find_element(By.CLASS_NAME, "a-size-base.a-color-base")
            ratings_text = ratings_element.text
            ratings_screen_protector.append(ratings_text + " out of 5 stars")
        except NoSuchElementException:
            ratings_screen_protector.append("No Rating")

except Exception as e:
    print(f"An error occurred: {str(e)}")
```
